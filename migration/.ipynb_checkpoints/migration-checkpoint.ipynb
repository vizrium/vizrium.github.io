```python
import pandas as pd
import csv

airport = pd.read_csv("airports.csv")
flight = pd.read_csv("flights-airport.csv")

newdict = {}
for x in range(len(airport)):
    newdict[airport.ix[x][0]] = airport.ix[x][7]

population = []
for x in flight['origin']:
    population.append( newdict[x])
```

```python
import re

def parseFile(filename, pattern, number):  
    myFile = open(filename,'r')
    result = []
    
    for line in myFile:
        x = re.sub("[,.;]", '', line).split()
        
        for word in x:
            mark = ''
            flag = 0
            count = 0
            
            for letter in word.lower():
                if letter not in pattern:
                    if len(mark) != 0 and letter != mark:
                        flag = 1
                        break
                    else:
                        mark = letter
                else:
                    count += 1
                    
            if flag == 0 and mark != '' and count > 0:
                result.append(word.lower())  
                
    return result
    
def getUniqueSorted(words):
    return sorted(set(res), key=len)

def getMostCommon(words):
    return max(set(words), key=words.count)
    
    
if __name__ == "__main__":
    
    filename = 'textSample.txt'
    pattern = 'rstlnaeiou'
    appear = 1
    
    res = parseFile(filename, pattern, appear)
    
    f = open('output1', 'w')
    f.write(str(getUniqueSorted(res)) + '\n')
    f.write(str(getMostCommon(res)))
    f.close()
```

```python
import random

def getRandomNumbers(start, end, count):
    return  random.sample(range(start, end+1), count)

def mergeThree(nums1, nums2, nums3):
    
    nums = nums1 + nums2 + nums3
    return sorted(set(nums))

def replaceNum(nums, num1, word1, num2, word2, word):
    
    for i in range(len(nums)):
        
        if nums[i]%num1 == 0 and nums[i]%num2 == 0:
            nums[i] = word
        
        elif nums[i]%num1 == 0:
            nums[i] = word1
        
        elif nums[i]%num2 == 0:
            nums[i] = word2

if __name__ == "__main__":
    
    start = 1
    end = 1000
    count = 20
    
    nums1 = getRandomNumbers(start, end, count)
    nums2 = getRandomNumbers(start, end, count)
    nums3 = getRandomNumbers(start, end, count)
    
    f = open('output2', 'w')
    f.write(str(nums1) + '\n')
    f.write(str(nums2) + '\n')
    f.write(str(nums3) + '\n')
    
    nums = mergeThree(nums1, nums2, nums3)
    f.write(str(nums) + '\n')
    replaceNum(nums, 2, 'Two', 5, 'Five', 'grapefruit')
    
    f.write(str(nums))
    f.close()
```
